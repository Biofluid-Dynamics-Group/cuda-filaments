```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt

# Parse the coordinates string
coords_str = "-0.425325 -3.71831e-08 0.262866 -0.262866 -0.425325 -2.18557e-08 -0.262866 0.425325 -2.18557e-08 -1.14902e-08 -0.262866 -0.425325 -1.14902e-08 -0.262866 0.425325 -1.14902e-08 0.262866 -0.425325 -1.14902e-08 0.262866 0.425325 0.262866 -0.425325 -2.18557e-08 0.262866 0.425325 -2.18557e-08 0.425325 0 -0.262866"
coords = np.array([float(x) for x in coords_str.split()])

# Reshape into triplets (x, y, z)
points = coords[:3 * (len(coords) // 3)].reshape(-1, 3)

# Create 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the points
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='r', marker='o')

# Connect points with lines
ax.plot(points[:, 0], points[:, 1], points[:, 2], 'b-')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Coordinates Plot')

# Equal aspect ratio
ax.set_box_aspect([1, 1, 1])

plt.show()
```
