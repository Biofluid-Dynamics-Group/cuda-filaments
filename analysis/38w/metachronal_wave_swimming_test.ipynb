```python
import numpy as np

file_path = '../data/metachronal_wave_swimming_test/20250225/ciliate_309fil_9000blob_8.00R_0.0010torsion_0.2182tilt_0.3000f_eff_1.4960theta0_0.0000freqshift_true_states.dat'
data = np.loadtxt(file_path)
print(data)
```

```python
data = np.delete(data, 1, axis=1)
print(data)
```

```python
import matplotlib.pyplot as plt

# Extract time array and phase time-series
time = data[:, 0]
phases = data[:, 1:]

# Plot all phase time-series
plt.figure(figsize=(10, 6))
for i in range(phases.shape[1]):
  plt.plot(time, phases[:, i], label=f'Cilium {i+1}')

plt.xlabel('Time')
plt.ylabel('Phase')
plt.title('Phase Time-Series for Each Cilium')
plt.show()

real_time = np.linspace(0, 1, len(time))

# Estimate the slope of every curve
slopes = []
for i in range(phases.shape[1]):
  slope = np.polyfit(real_time, phases[:, i], 1)[0]
  slopes.append(slope)
  print(f'Slope of Cilium {i+1}: {slope}')

# Verify that all slopes are approximately 2pi
all_slopes_same = all(np.isclose(slope, 2 * np.pi) for slope in slopes)
print(f'All slopes are approximately 2pi: {all_slopes_same}')
```
