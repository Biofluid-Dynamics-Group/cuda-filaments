```python
import numpy as np
import matplotlib.pyplot as plt


path = '../data/fixed_body_emergent_rand_0/20250401/ciliate_309fil_9000blob_8.00R_0.0100torsion_0.2182tilt_0.3000f_eff_1.4960theta0_0.0000freqshift_true_states.dat'

# Load the data from the .dat file
data = np.loadtxt(path)

# Extract time values
time = data[:, 0] / 500.0  # Since i/500 is time

# Extract psi1 values for all cilia (columns 2 to 2+309-1)
psi1_values = data[:, 2:2+309]

# Create a figure and axes
plt.figure(figsize=(12, 6))

# Plot each psi1 value over time without a legend
for i in range(psi1_values.shape[1]):
  plt.plot(time[:200], psi1_values[:200, i])

plt.xlabel('Time')
plt.ylabel('psi1 values')
plt.title('Phase of Cilia (psi1) over Time')

plt.tight_layout()
plt.show()
```
