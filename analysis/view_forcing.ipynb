```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
# Define file paths
q1_file = "../input/forcing/platy_reference_phase_generalised_forces_NSEG=20_SEP=2.600000_FEFF=0.300000_THETA0=1.495996.dat"
q2_file = "../input/forcing/platy_reference_angle_generalised_forces_NSEG=20_SEP=2.600000_FEFF=0.300000_THETA0=1.495996.dat"

# Read the first generalised forces function (q1)
with open(q1_file, 'r') as f:
  data = f.read().strip().split()
  n_samples = int(data[0])
  forces_1 = np.array([float(val) for val in data[1:]])

# Read the second generalised forces function (q2)
with open(q2_file, 'r') as f:
  data = f.read().strip().split()
  n_samples_2 = int(data[0])
  forces_2 = np.array([float(val) for val in data[1:]])

# Generate evenly spaced s values from 0 to 2π
psi_values = np.linspace(0, 2*np.pi, n_samples)

# Create figure and main axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot q1 and q2 on main axes
ax.plot(psi_values, forces_1, label=r'$q_1$')
ax.plot(psi_values, forces_2, label=r'$q_2$')
ax.set_xlabel('phase')
ax.set_ylabel('')
ax.set_title('forcing')
ax.legend()
ax.grid(True)

# Create inset axes for the zoomed region
# axins = inset_axes(ax, width="40%", height="40%", loc="right")

# Plot the same data in the inset
axins.plot(psi_values, forces_1, label=r'$q_1$')
axins.plot(psi_values, forces_2, label=r'$q_2$')
axins.grid(True)

# Set the limits for the zoomed region
x_min = 0.3*2*np.pi - 0.5
x_max = 0.3*2*np.pi + 0.5
axins.set_xlim(x_min, x_max)

plt.show()
```
